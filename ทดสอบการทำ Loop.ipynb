```python
#vectorized version

import numpy as np
import time

a = np.random.rand(1000000)
b = np.random.rand(1000000)
```

```python
# for loop
c = 0
tic = time.time()
for i in range(1000000):
    c += a[i]*b[i]
toc = time. time()

print (c)
print ("for loop:"+ str(1000*(tic-toc))+"ms")

# vectorized version
tic = time.time()
c = np.dot(a,b)
toc = time.time()

print (c)
print ("vectorized versoin:"+ str(1000*(tic-toc))+"ms")
```
