```python
import numpy as np
```

```python
# การ broadcasting ที่อยู่ใน Library Numpy นั้นมีไว้เพื่อสำหรับใช้ในการคูญ Matrix หรือ Array ที่ Shape(ขนาดไม่เท่ากัน) ไม่เท่ากัน
```

```python
A = np.array([[56.0,0.0,4.4,68.0],[1.2,104.0,52.0,8.0],[1.8,135.0,99.0,0.9]])

print (A)
```

```python
cal  = A.sum(axis=0)
print (cal)
```

```python
percentage = 100*A/cal.reshape(1,4) # ต้อง reshape ตลอด เพืื่อ Bug prevention
print (percentage)
```
