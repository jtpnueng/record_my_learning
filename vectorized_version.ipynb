```python
import numpy as np
import time
```

```python
# 1 fot loop
a = np.random.rand(1000000)
b = np.random.rand(1000000)

c = 0
tic = time.time()
for i in range(1000000):
    c += a[i]*b[i]
toc = time.time()

print (c)
print ("for loop:"+ str(1000*(tic-toc))+"ms")

tic = time.time()
c = np.dot(a,b)
toc = time.time()

print (c)
print ("vectorized version:" + str(1000*(tic-toc))+"ms")
```
